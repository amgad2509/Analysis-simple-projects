```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
df=pd.read_csv('auto-mpg.csv')
df
```

```python
df.head()
```

```python
sns.relplot(x='cylinders',y='displacement',data=df,kind='line')   #rel between cyl and displacement(directly proportional)
```

```python
sns.catplot(x='origin',data=df,kind='count')
```

```python
sns.relplot(x='cylinders',y='displacement',data=df,col='origin',hue='origin')
```

```python
sns.catplot(x='model year',y='mpg',data=df,kind='box')   #As we can see From the model 80, the median for mpg is more than the models before it.
```

```python
sns.relplot(x ="weight", y ="mpg",data=df,kind='scatter') #mpg is increase when the weight is less
```

```python
plt.figure(figsize=(12,10))
cor = df.corr()                       #display co-relation between all columns
sns.heatmap(cor,annot=True)
```

```python
sns.distplot(df['displacement'],kde=False,bins=15)
```

```python
sns.pairplot(df)
```

```python
df1=df.sample(50)
df1
```

```python
sns.relplot(x='acceleration',y='mpg',data=df1)
```

```python
sns.relplot(x='displacement',y='mpg',data=df) #with low mpg ,,, high displacement
```

```python
sns.relplot(x='weight',y='displacement',data=df) #weight increase.... displacement increase
```
